```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
```

```python
def errors(L1, L2,w1, w2 ):
    first  = (1.0/2)*L1*(w1**2)
    second = (1.0/2)*L2*(w2**2)    
    return (first + second)
```

## Visualization of loss function

```python
L2 = 1
L1 = 10*L2

errors_list = []    
for i in range(-10,10,1):
    for j in range(-10,10,1):
        w1 = i
        w2 = j
        error = errors(L1, L2, w1, w2)
        errors_list.append(error)

x = np.linspace(-10,15,len(errors_list))

plt.plot(x,errors_list)
```

The mentioned error function has many local minima and one maxima in a given range of weights

## Gradient Descent algorithm

```python
threshold = 0.5
lr = 0.01

error_list = []
L1 = 1
L2 = 10*L1

errors_list = []
w1 = 0
w2 = 1

for i in range(5):
    errors_list = []
    for j in range(15):  
        error = errors(L1, L2, w1, w2)
        errors_list.append(error)
        w1 = w1 - lr*(L1*w1)
        w2 = w2 - lr*(L2*w2)
    error_list.append(np.sum(errors_list))   
    print("error = ", np.sum(errors_list))
```
